```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 1000)
```

```python
filepath = "/Users/laurazhang/Documents/Thesis/"
race_raw = pd.read_csv(filepath + '/raw/nhgis/nhgis0001_ts_nominal_tract.csv')
inc_raw = pd.read_csv(filepath + '/raw/nhgis/nhgis0003_ts_nominal_tract.csv')
```

```python
inc_df = inc_raw.iloc[:, np.r_[1:3, 12:41]]
inc_df.head()
```

```python
keep_racecols = ['NHGISCODE', 'GISJOIN', 'YEAR', 'STATE', 'STATEFP', 'STATENH', 'COUNTY', 'COUNTYFP', 'COUNTYNH', 'TRACTA'] \
    + ['AV0AA', 'B18AA', 'B18AB', 'B18AC', 'B18AD', 'B18AE', 'B08AA', 'B08AB', 'B08AC', 'B08AD', 'B08AE', 'B08AF', 'B08AG', 'B08AH', 'B08AI', 'B08AJ']
race_df = race_raw[keep_racecols]
```

```python
rename_racecols = {'AV0AA': 'totpop',
                'B18AA': 'countwhite',
                'B18AB': 'countblack',
                'B18AC': 'countnative',
                'B18AD': 'countasian',
                'B18AE': 'count2more',
                'B08AA': 'dcountwhite',
                'B08AB': 'dcountblack',
                'B08AC': 'dcountnative',
                'B08AD': 'dcountjap',
                'B08AE': 'dcountchin',
                'B08AF': 'dcountfilip',
                'B08AG': 'dcountkorean',
                'B08AH': 'dcountapi',
                'B08AI': 'dcountother',
                'B08AJ': 'dcount2more'}
race_df = race_df.rename(columns=rename_racecols)
rename_inccols = {'B71AA': 'hhinc10k',
                'B71AB': 'hhinc15k',
                'B71AC': 'hhinc20k',
                'B71AD': 'hhinc25k',
                'B71AE': 'hhinc30k',
                'B71AF': 'hhinc35k',
                'B71AG': 'hhinc40k',
                'B71AH': 'hhinc45k',
                'B71AI': 'hhinc50k',
                'B71AJ': 'hhinc60k',
                'B71AK': 'hhinc75k',
                'B71AL': 'hhinc100k',
                'B71AM': 'hhinc125k',
                'B71AN': 'hhinc150k',
                'B71AO': 'hhinc150+',
                'A90AA': 'finc10k',
                'A90AB': 'finc15k',
                'A90AC': 'finc20k',
                'A90AD': 'finc25k',
                'A90AE': 'finc30k',
                'A90AF': 'finc35k',
                'A90AG': 'finc40k',
                'A90AH': 'finc45k',
                'A90AI': 'finc50k',
                'A90AJ': 'finc60k',
                'A90AK': 'finc75k',
                'A90AL': 'finc100k',
                'A90AM': 'finc125k',
                'A90AN': 'finc150k',
                'A90AO': 'finc150+'}
inc_df = inc_df.rename(columns = rename_inccols)
```

```python
inc_df.head()
```

```python
clean_df = pd.merge(race_df, inc_df, on=['GISJOIN', 'YEAR'], how='left')
```

```python
clean_df = clean_df[clean_df['YEAR'] >= 1990]
```
